```python
# Install dependencies
!pip install python-docx pandas matplotlib

# Import libraries
from google.colab import files
import pandas as pd
from docx import Document
import io

# Upload the Word file
uploaded = files.upload()  # Select your 'Incubation final tables.docx'

# Load the document
filename = list(uploaded.keys())[0]
doc = Document(io.BytesIO(uploaded[filename]))

# Function to extract all tables into DataFrames
def extract_tables_from_docx(doc):
    all_tables = []
    for table in doc.tables:
        data = []
        for row in table.rows:
            row_data = [cell.text.strip() for cell in row.cells]
            data.append(row_data)
        df = pd.DataFrame(data)
        all_tables.append(df)
    return all_tables

tables = extract_tables_from_docx(doc)
print(f"Extracted {len(tables)} tables")
tables[0].head()  # Display first table
```

```python
for i, df in enumerate(tables):
    print(f"Table {i+1}:")
    display(df.head())
    print(f"Shape of Table {i+1}: {df.shape}\n")
```

```python
# 1. Select the first table
df_selected_table = tables[0].copy()

# 2. and 3. Inspect header and create new column names
# Assuming the first two rows form the header
new_columns = []
for col in df_selected_table.columns:
    param = df_selected_table.iloc[0, col]
    time = df_selected_table.iloc[1, col]

    # Safely get stripped values, replacing None with empty string
    param_str = param.strip() if isinstance(param, str) else ''
    time_str = time.strip() if isinstance(time, str) else ''

    # Handle the first column which is just treatment
    if col == 0:
        new_columns.append('Treatment')
    elif param_str and time_str:
        new_columns.append(f'{param_str} - {time_str}')
    elif param_str:
         new_columns.append(param_str) # Handle cases where only parameter is present
    else:
        new_columns.append(f'Col_{col}') # Placeholder for unhandled cases or None columns

# 4. Assign new column names and skip original header rows and statistical rows
df_selected_table.columns = new_columns
# Exclude header rows (0 and 1) and rows containing 'S. Em', 'CD @ 1%', and 'Initial'
df_selected_table = df_selected_table.iloc[2:]
df_selected_table = df_selected_table[~df_selected_table['Treatment'].isin(['S. Em', 'CD @ 1%', 'Initial'])].reset_index(drop=True)


# 5. Drop columns that are entirely 'None' as identified in header inspection
# Based on the output, the last two columns appear to be 'None' placeholders ('Col_16', 'Col_17')
df_selected_table = df_selected_table.drop(columns=['Col_16', 'Col_17'], errors='ignore')


# 7. Reshape the DataFrame from wide to long format
# Identify the columns to melt - all except 'Treatment'
id_vars = ['Treatment']
value_vars = [col for col in df_selected_table.columns if col not in id_vars]

df_long = pd.melt(df_selected_table,
                  id_vars=id_vars,
                  value_vars=value_vars,
                  var_name='Parameter_Time',
                  value_name='Value')

# 8. Extract parameter and DAI from 'Parameter_Time'
# Split only on the first occurrence of ' - ' to handle cases like 'pH - 0 DAI'
df_long[['Parameter', 'DAI_str']] = df_long['Parameter_Time'].str.split(' - ', n=1, expand=True)

# Clean up the DAI string to extract the number
df_long['DAI'] = df_long['DAI_str'].str.replace(r'\s?DAI', '', regex=True).str.replace(r'\n', '', regex=True)

# 6. Convert the 'Value' and 'DAI' columns to numeric
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')
df_long['DAI'] = pd.to_numeric(df_long['DAI'], errors='coerce')


# Drop the intermediate columns
df_long = df_long.drop(columns=['Parameter_Time', 'DAI_str'], errors='ignore')

# Drop rows where 'Value' or 'DAI' could not be converted to numeric (NaN)
df_long.dropna(subset=['Value', 'DAI'], inplace=True)

# 9. Pivot the DataFrame
df_pivot = df_long.pivot_table(index=['Treatment', 'DAI'], columns='Parameter', values='Value').reset_index()

display(df_pivot.head())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# 2. Define a function create_sequences
def create_sequences(data, seq_length, target_name):
    X, y = [], []
    for i in range(len(data) - seq_length):
        # Extract features (all columns except the target) for the current sequence
        seq_features = data.iloc[i:(i + seq_length)].drop(columns=[target_name])
        # Extract the target value for the next step
        seq_target = data.iloc[i + seq_length][target_name]

        X.append(seq_features.values)
        y.append(seq_target)

    return np.array(X), np.array(y)

# 3. Identify unique Treatment and Parameter values
unique_treatments = df_long['Treatment'].unique()
unique_parameters = df_long['Parameter'].unique()

# Lists to store sequences and targets from all combinations
all_X, all_y = [], []
all_treatments, all_parameters = [], []

# Define sequence length
sequence_length = 3 # Example sequence length

# 4. Iterate through each unique Treatment and Parameter combination
for treatment in unique_treatments:
    for parameter in unique_parameters:
        # Filter df_long for the current combination
        df_combination = df_long[(df_long['Treatment'] == treatment) & (df_long['Parameter'] == parameter)].copy()

        # Sort by DAI
        df_combination = df_combination.sort_values(by='DAI')

        # Handle potential missing values in 'Value' by interpolating
        # This is one approach; another could be dropping rows or using forward/backward fill
        df_combination['Value'] = df_combination['Value'].interpolate(method='linear')

        # Drop rows with any remaining NaNs (e.g., if interpolation wasn't possible at the ends)
        df_combination.dropna(inplace=True)


        # Check if there is enough data to create sequences
        if len(df_combination) >= sequence_length + 1: # Need at least seq_length + 1 data points
            # Create sequences for this combination
            X_seq, y_seq = create_sequences(df_combination, sequence_length, 'Value')

            # Store the created sequences and targets
            all_X.append(X_seq)
            all_y.append(y_seq)

            # Store corresponding Treatment and Parameter for each sequence
            # Each sequence generated from df_combination will have the same treatment and parameter
            all_treatments.extend([treatment] * len(X_seq))
            all_parameters.extend([parameter] * len(X_seq))

# 5. Combine all the sequences and targets into single NumPy arrays
# Need to reshape the sequences to be 2D for combining
# Also need to handle the feature columns; currently, create_sequences gives all columns except target
# Let's modify create_sequences or process X_seq here to only include relevant features for LSTM
# For this task, let's assume the features are 'DAI' and the 'Value' from previous steps.
# The 'Treatment' and 'Parameter' are being handled separately for tracking.

# Re-defining create_sequences to explicitly select features
def create_sequences_features(data, seq_length, target_name, feature_names):
    X, y = [], []
    for i in range(len(data) - seq_length):
        # Extract features for the current sequence
        seq_features = data.iloc[i:(i + seq_length)][feature_names]
        # Extract the target value for the next step
        seq_target = data.iloc[i + seq_length][target_name]

        X.append(seq_features.values)
        y.append(seq_target)

    return np.array(X), np.array(y)

# Features to use for the LSTM model
feature_names = ['DAI', 'Value'] # Using DAI and the 'Value' from previous time steps

all_X, all_y = [], []
all_treatments, all_parameters = [], []

for treatment in unique_treatments:
    for parameter in unique_parameters:
        df_combination = df_long[(df_long['Treatment'] == treatment) & (df_long['Parameter'] == parameter)].copy()
        df_combination = df_combination.sort_values(by='DAI')
        df_combination['Value'] = df_combination['Value'].interpolate(method='linear')
        df_combination.dropna(inplace=True)

        if len(df_combination) >= sequence_length + 1:
            X_seq, y_seq = create_sequences_features(df_combination, sequence_length, 'Value', feature_names)
            all_X.append(X_seq)
            all_y.append(y_seq)
            all_treatments.extend([treatment] * len(X_seq))
            all_parameters.extend([parameter] * len(X_seq))

# Concatenate the lists of arrays
X = np.concatenate(all_X, axis=0) if all_X else np.array([])
y = np.concatenate(all_y, axis=0) if all_y else np.array([])
treatments_array = np.array(all_treatments)
parameters_array = np.array(all_parameters)


# 6. Split the combined data into training and testing sets
# We can also stratify by treatment and parameter if needed, but for a simple split:
X_train, X_test, y_train, y_test, treatments_train, treatments_test, parameters_train, parameters_test = train_test_split(
    X, y, treatments_array, parameters_array, test_size=0.2, random_state=42
)

print(f"Original data shape (X): {X.shape}, (y): {y.shape}")
print(f"Training data shape (X_train): {X_train.shape}, (y_train): {y_train.shape}")
print(f"Testing data shape (X_test): {X_test.shape}, (y_test): {y_test.shape}")
print(f"Training treatments shape: {treatments_train.shape}, Training parameters shape: {parameters_train.shape}")
print(f"Testing treatments shape: {treatments_test.shape}, Testing parameters shape: {parameters_test.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Define the LSTM model architecture
model = Sequential()

# Add an LSTM layer
# The input shape is (sequence_length, number_of_features)
# X_train.shape[1] is the sequence_length, X_train.shape[2] is the number_of_features
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))

# Add Dense layers
model.add(Dense(units=25, activation='relu'))

# Add the output layer
model.add(Dense(units=1))

# Print the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Print completion message
print("LSTM model training completed.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# 1. Use the trained model to make predictions on the test set
lstm_predictions = model.predict(X_test)

# Reshape predictions to be 1D if necessary
lstm_predictions = lstm_predictions.flatten()

# 2. Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, lstm_predictions)
print(f"Mean Squared Error (MSE): {mse:.4f}")

# 3. Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")

# 4. Create a scatter plot comparing the actual values and the predicted values
plt.figure(figsize=(8, 8))
plt.scatter(y_test, lstm_predictions, alpha=0.5)

# 5. Add a diagonal line representing perfect predictions
max_val = max(y_test.max(), lstm_predictions.max())
min_val = min(y_test.min(), lstm_predictions.min())
plt.plot([min_val, max_val], [min_val, max_val], 'k--', lw=2) # k-- for black dashed line, lw for line width

# 6. Add a title and labels to the scatter plot
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Actual vs. Predicted Values (LSTM Model)")

# 7. Display the scatter plot
plt.show()

# 8. Plot the training and validation loss over epochs
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('LSTM Model Training and Validation Loss over Epochs')
plt.legend()
plt.show()
```
